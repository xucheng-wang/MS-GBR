```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv("../result/EM-f-BSG2/EM-f-BSG.csv")

fig, ax = plt.subplots()

# Plot x_ne_L curve with shaded area
ax.plot(data['mean_x_ne_L'], label='x_ne_L', linewidth=1, linestyle='-')
ax.fill_between(range(len(data)), data['lower_quartile_x_ne_L'], data['upper_quartile_x_ne_L'], alpha=0.3)

# Plot x_hat_L curve with shaded area
ax.plot(data['mean_x_hat_L'], label='x_hat_L', linewidth=1, linestyle='-')
ax.fill_between(range(len(data)), data['lower_quartile_x_hat_L'], data['upper_quartile_x_hat_L'], alpha=0.3)

# Add labels, title, and legend
ax.set_xlabel('Iteration')
ax.set_ylabel('Regret')
ax.set_title('email-fully-homophily')
ax.legend()

plt.show()
```
